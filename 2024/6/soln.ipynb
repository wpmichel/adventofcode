```python
NEWLINE = '\n'
GUARD = '^'
OBSTACLE = '#'
def parse_input(string: str): 
    string = string.lstrip(NEWLINE).rstrip(NEWLINE)
    arr = [list(col) for col in string.split(NEWLINE)] 
    pos = None
    for row in range(len(arr)): 
        for col in range(len(arr[0])): 
            if arr[row][col] == GUARD: 
                pos = (row, col)
    return arr, pos

    
```

```python
# puzzle 1 
test_input = """
....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#...
"""
def track_path(arr: list[list[str]], row: int, col: int):
    def get_offsets(direction): 
        UP, RIGHT, DOWN, LEFT = 0, 1, 2, 3
        if direction == UP:
            return -1, 0
        elif direction == RIGHT:
            return 0, 1
        elif direction == DOWN:
            return 1, 0
        else:
            return 0, -1

    direction = 0
    visited = set()
    
    while True: 
        visited.add((row, col))
        row_offset, col_offset = get_offsets(direction)
        
        row_next, col_next = row+row_offset, col+col_offset
        if not (0 <= row_next <= len(arr) - 1 and 0 <= col_next <= len(arr[0]) - 1):
            break 

        if arr[row_next][col_next] == OBSTACLE: 
            direction = (direction+1) % 4
            row_offset, col_offset = get_offsets(direction)
            row_next, col_next = row+row_offset, col+col_offset
        
        row, col = row_next, col_next
    return visited


arr, (row, col) = parse_input(puzzle_input)
visited = track_path(arr, row, col)
print(len(visited), visited)
```

```python
# puzzle 2

def detect_loop(arr: list[list[str]], row: int, col: int):
    def get_offsets(direction): 
        UP, RIGHT, DOWN, LEFT = 0, 1, 2, 3
        if direction == UP:
            return -1, 0
        elif direction == RIGHT:
            return 0, 1
        elif direction == DOWN:
            return 1, 0
        else:
            return 0, -1

    direction = 0
    visited = set()
    while True: 
        if (row, col, direction) in visited: 
            return True
            

        visited.add((row, col, direction))
        row_offset, col_offset = get_offsets(direction)
        
        row_next, col_next = row+row_offset, col+col_offset
        if (not 0 <= row_next <= len(arr) - 1) or (not 0 <= col_next <= len(arr[0]) - 1):
            return False

        if arr[row_next][col_next] == OBSTACLE: # keep turning
            direction = (direction+1) % 4
        else:
            row, col = row_next, col_next

test_input = """
....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#...
"""

arr, (starting_row, starting_col) = parse_input(puzzle_input)
loops = set()
for row in range(len(arr)):
    for col in range(len(arr[0])): 
        if (0 <= row <= len(arr) - 1 and 0 <= row <= len(arr[0]) - 1): 
            before, arr[row][col] = arr[row][col], OBSTACLE
            if detect_loop(arr, row, col): 
                loops.add((row, col))
            arr[row][col] = before
if (starting_row, starting_col) in loops: 
    loops.remove((starting_row, starting_col))
print(len(loops), loops)

```
