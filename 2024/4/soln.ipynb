```python

def input_to_2d_arr(string: str): 
    lines = string.lstrip('\n').rstrip('\n').split('\n')
    return [list(l) for l in lines]
```

```python
# puzzle 1
test_input = """
MMMSXXMASM
MSAMXMSMSA
AMXSXMAAMM
MSAMASMSMX
XMASAMXAMM
XXAMMXXAMA
SMSMSASXSS
SAXAMASAAA
MAMMMXMMMM
MXMXAXMASX
"""
# Intuition: we only really need to check to the right, down, and down diagonals because we are searching for XMAS and the backwards SAMX
def find_xmas(arr): 
    found = []
    words = ["XMAS", "SAMX"]
    rows, cols, word_pos = len(arr), len(arr[0]), range(len(words[0]))
   
    for row in range(rows):
        for col in range(cols):
            try:
                space_down = False
                if row + 3 < rows: # down
                    space_down = True
                    word = ''.join([arr[row+offset][col] for offset in word_pos])
                    if word in words: 
                        found.append((word, (row, col, "down")))
                

                if col + 3 < cols: # right 
                    word = ''.join([arr[row][col+offset] for offset in word_pos])
                    if word in words: 
                        found.append((word, (row, col, "right")))
            
                    if space_down: # diagonal, down, right 
                        word = ''.join([arr[row+offset][col+offset] for offset in word_pos])
                        if word in words: 
                            found.append((word, (row, col, "down_right")))

                
                if col - 3 >= 0 and space_down: # diagonal, down, left 
                    word = ''.join([arr[row+offset][col-offset] for offset in word_pos])
                    if word in words: 
                        found.append((word, (row, col, "down_left")))
            except Exception as e: 
                print(f"Failed at {row}, {col}")
                raise e

                
    return found

print(find_xmas(input_to_2d_arr(test_input)))
print(len(find_xmas(input_to_2d_arr(test_input))))

print(find_xmas(input_to_2d_arr(puzzle_input)))
print(len(find_xmas(input_to_2d_arr(puzzle_input))))
```

```python
# puzzle 2


def find_x_mas(arr):
    found = []
    rows, cols = len(arr), len(arr[0])
   
    for row in range(1, rows - 1):
        for col in range(1, cols - 1):
            try:
                if arr[row][col] == 'A': 
                    if (
                        (
                            (arr[row-1][col-1] == 'M' and arr[row+1][col+1] == 'S') or 
                            (arr[row-1][col-1] == 'S' and arr[row+1][col+1] == 'M')
                        ) and 
                        (
                            (arr[row-1][col+1] == 'M' and arr[row+1][col-1] == 'S') or 
                            (arr[row-1][col+1] == 'S' and arr[row+1][col-1] == 'M')
                        )
                    ):
                        found.append((row, col))
                
            except Exception as e: 
                print(f"Failed at {row}, {col}")
                raise e

                
    return found

print(find_x_mas(input_to_2d_arr(test_input)))
print(len(find_x_mas(input_to_2d_arr(test_input))))
print(find_x_mas(input_to_2d_arr(puzzle_input)))
print(len(find_x_mas(input_to_2d_arr(puzzle_input))))
```
