```python
from collections import defaultdict


def parse_input(string): 
    string = string.lstrip('\n').rstrip('\n')
    rules = True 
    lines = iter(string.split('\n'))
    rules = defaultdict(set)
    updates = []
    while True: 
        line = next(lines)
        if line == "": 
            break

        before, after = line.split('|')
        rules[after].add(before)
    
    while True: 
        line = next(lines, None)
        if line is None: 
            break
        updates.append(line.split(','))
        
    return rules, updates

print(parse_input(test_input))
```

```python
# puzzle 2
def is_compliant_update(rules: dict[set], update, seen): 
    constraints: set = rules[update]
    # If there is an intersection between constraints and seen, then we have violated the rules
    for s in seen: 
        if s in constraints: 
            return False
    return True

def is_compliant_updates(rules: dict[set], updates): 
    reversed_updates = updates[::-1]
    seen = set() 
    compliant = True
    for update in reversed_updates: 
        seen.add(update)
        compliant = is_compliant_update(rules, update, seen)
        if not compliant: 
            return False
    return True 

def find_uncompliant_updates(rules: dict[set], manuals): 
    uncompliant_updates = []
    for updates in manuals: 
        if not is_compliant_updates(rules, updates): 
            uncompliant_updates.append(updates)        
    return uncompliant_updates

def solve(state, get_candidates, is_valid, is_complete):
    """
    Generic backtracking solver.
    
    Parameters:
    state: Current partial solution
    get_candidates: Function that returns possible values for next position
    is_valid: Function that checks if state remains valid after adding a value
    is_complete: Function that checks if solution is complete
    """
    if is_complete(state):
        return state
        
    for candidate in get_candidates(state):
        state.append(candidate)
        if is_valid(state):
            result = solve(state, get_candidates, is_valid, is_complete)
            if result is not None:
                return result
        state.pop()
    
    return None

def reorder_update(rules, updates): 
    def get_candidates(state):
        return set(updates).difference(set(state))
        
    def is_valid(state):
        return is_compliant_updates(rules, state)
        
    def is_complete(state):
        return len(state) == len(updates)
    
    return solve([], get_candidates, is_valid, is_complete)

def compute_middle_sum(updates): 
    running_sum = 0 
    for update in updates: 
        if len(update) % 2 == 0: 
            raise Exception("even")
        running_sum += int(update[len(update) // 2])
    return running_sum


for rules, manuals in [parse_input(s) for s in [test_input, puzzle_input]]: 
    updates = find_uncompliant_updates(rules, manuals)
    print("got bad updates")
    reordered = []
    for idx, update in enumerate(updates): 
        print(f"working on {idx}/{len(updates)}")
        reordered.append(reorder_update(rules, update))
    print(compute_middle_sum(reordered))

```

```python
from collections import defaultdict, deque

def build_graph(rules, updates):
    # Build adjacency list and in-degree count for each node
    graph = defaultdict(list)
    in_degree = defaultdict(int)
    
    for update in updates:
        update_set = set(update)
        for node in update:
            for prereq in rules[node]:
                if prereq in update_set:
                    graph[prereq].append(node)
                    in_degree[node] += 1
    return graph, in_degree

def topological_sort(nodes, graph, in_degree):
    result = []
    queue = deque([n for n in nodes if in_degree[n] == 0])
    
    while queue:
        node = queue.popleft()
        result.append(node)
        
        for neighbor in graph[node]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)
                
    return result if len(result) == len(nodes) else None

def reorder_update(rules, update):
    graph, in_degree = build_graph(rules, [update])
    return topological_sort(update, graph, in_degree)

def find_and_reorder_updates(rules, manuals):
    reordered = []
    for update in manuals:
        graph, in_degree = build_graph(rules, [update])
        orig_order = topological_sort(update, graph, in_degree)
        if orig_order != update:
            fixed = reorder_update(rules, update)
            if fixed:
                reordered.append(fixed)
    return reordered
rules, manuals = parse_input(puzzle_input)
middle_sum = sum(int(update[len(update)//2]) for update in find_and_reorder_updates(rules, manuals))
print(middle_sum)
```
