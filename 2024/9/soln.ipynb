```python
test_input = """2333133121414131402"""

def calc_checksum(string): 
    """
    >>> calc_checksum("0099811188827773336446555566")
    1928
    >>> calc_checksum("00992111777.44.333....5555.6666.....8888..")
    2858
    """
    total = 0
    for idx, val in enumerate(string): 
        if val != '.':
            total += (idx*int(val))
    return total


doctest.run_docstring_examples(calc_checksum, None)


def compact(string): 
    """
    >>> ''.join(map(str, compact("12345")))
    '022111222'
    """
    string = f"{string}0" # denote no free space
    file_size_pos, space_pos, file_num = 0, 1, 0
    metadata = {}
    while space_pos < len(string):
        file_size, space = int(string[file_size_pos]), int(string[space_pos])
        metadata[file_num] = (file_size, space)
        file_num, file_size_pos, space_pos = file_num+1, file_size_pos+2, space_pos+2
    
    compacted = []
    start, end = 0, file_num-1
    while start <= end: 
        compacted.extend([start]*metadata[start][0]) # add file block
        metadata[start] = (0, metadata[start][1])
        
        if start != end:
            for _ in range(int(metadata[start][1])): # fill free space
                if metadata[end][0] <= 0: 
                    end -= 1
                if start != end:
                    compacted.append(end)
                    metadata[end] = (metadata[end][0] - 1, metadata[end][1])
        start += 1
    
    return compacted

doctest.run_docstring_examples(compact, None)


#blocks = compact(test_input)
blocks = compact(puzzle_input)
print(calc_checksum(blocks), blocks)
```
