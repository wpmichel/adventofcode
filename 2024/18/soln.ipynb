# Puzzle 1

```python
from heapq import heappop, heappush
BITE = '#'

def build_arr(height, width, bites): 
    arr = [['.']*(width+1) for _ in range(height+1)]
    for (y,x) in bites: 
        arr[y][x] = '#'
    return arr

def shortest_path(arr, start, end): 
    def get_neighbors(arr, pos): 
        for (yo, xo) in [(-1, 0), (1, 0), (0, -1), (0, 1)]: 
            y2, x2 = pos[0] + yo, pos[1]+xo
            if in_bounds(arr, y2, x2) and arr[y2][x2] != BITE: 
                yield (y2,x2)

    # (cost, start, path)
    visited = {start: 0}
    queue = [(0, start)]
    while queue:
        cost, pos = heappop(queue)

        if pos == end: 
            return cost
        
        if pos in visited and cost > visited[pos]:
            continue

        for next_pos in get_neighbors(arr, pos): 
            next_cost = cost + 1
            if next_pos not in visited or next_cost  < visited[next_pos]: 
                visited[next_pos] = next_cost
                heappush(queue, (next_cost, next_pos))

    return -1

length, width = 6, 6
arr = build_arr(length, width, parse_input(test_input)[:12])
print(shortest_path(arr, (0, 0), (6, 6)))

length, width = 70, 70
arr = build_arr(length, width, parse_input(puzzle_input)[:1024])
print(shortest_path(arr, (0, 0), (length, width)))
```
