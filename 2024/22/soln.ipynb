```python
from collections import defaultdict

def get_price_changes(sellers): 
    prices = defaultdict(lambda: 0)
    for seller in sellers: 
        price_history = []
        last_price = None
        price = seller
        seen = set()
        for _ in range(2_000+1):
            ones = price % 10 
            if last_price is not None:
                price_history.append(ones - last_price)  # Current minus previous gives us the changes shown
                if len(price_history) >= 4:
                    history = tuple(price_history[-4:])  
                    if history not in seen: # only count first time we've seen for this seller
                        prices[history] += ones
                        seen.add(history)
            last_price = ones 
            price = advance(price)
    return prices


test_input = """
1
2
3
2024
"""
sellers = parse_input(puzzle_input)
#sellers = parse_input(test_input)
#sellers = [123]
prices = get_price_changes(sellers)
best_seq, best_bananas = None, 0
for seq, bananas in prices.items():
    if bananas > best_bananas: 
        best_seq = seq
        best_bananas = bananas
print(best_seq, best_bananas)
```
