```python
import doctest
from pprint import pprint

NEWLINE = '\n'
ROBOT_CHAR = '@'
BOX = 'O'
BARRIER = '#'
SPACE = '.'
UP, DOWN, LEFT, RIGHT = '^', 'v', '<', '>'

def parse_input(string: str): 
    string = string.strip(NEWLINE)
    warehouse, moves = string.split(NEWLINE+NEWLINE)
    warehouse = [list(line) for line in warehouse.split(NEWLINE)]
    moves = list(moves.replace(NEWLINE, ''))
    return warehouse, moves

doctest.run_docstring_examples(parse_input, None)
```

```python
test_input = """
########
#..O.O.#
##@.O..#
#...O..#
#.#.O..#
#...O..#
#......#
########

<^^>>>vv<v>>v<<
"""
MOVES = {
    UP: (-1, 0), 
    DOWN: (1, 0),
    LEFT: (0, -1), 
    RIGHT: (0, 1),
}


def shift(warehouse, push, move):
    """
    >>> warehouse = [list("#.@#")]
    >>> shift(warehouse, (0, 2), RIGHT)
    (0, 2)
    >>> warehouse  
    [['#', '.', '@', '#']]

    >>> warehouse = [list("#O@#")]
    >>> shift(warehouse, (0, 2), LEFT)
    (0, 2)
    >>> warehouse  
    [['#', 'O', '@', '#']]

    >>> warehouse = [list("#.O@#")]
    >>> shift(warehouse, (0, 3), LEFT)
    (0, 2)
    >>> warehouse  
    [['#', 'O', '@', '.', '#']]
    
    >>> warehouse = [list("#.OO@")]
    >>> shift(warehouse, (0, 4), LEFT)
    (0, 3)
    >>> warehouse  
    [['#', 'O', 'O', '@', '.']]

    >>> warehouse = [['#', '.', '@', 'O', '.', 'O', '.', '#']]
    >>> shift(warehouse, (0, 2), RIGHT)
    (0, 3)
    >>> warehouse  
    [['#', '.', '.', '@', 'O', 'O', '.', '#']]
    """ 
    dy, dx = MOVES[move]
    py, px =push[0], push[1]
    shoved = (py + dy, px + dx)
    sy, sx = shoved[0], shoved[1]
    if warehouse[sy][sx] == BARRIER: # barriers don't move
        return push
    
    if warehouse[sy][sx] == BOX: # Let's try to move it. If it can move, then the next space will be SPACE
        shift(warehouse, shoved, move)

    if warehouse[sy][sx] == SPACE: # open, can just shift
        warehouse[sy][sx], warehouse[py][px] =  warehouse[py][px], warehouse[sy][sx]
        return (sy, sx)
    return push

doctest.run_docstring_examples(shift, None)
    
def calculate_gps_coordinates(warehouse): 
    """
    >>> calculate_gps_coordinates([list("#######"), list("#...O.."), list("#......")])
    104
    """

    total = 0 
    for y, col in enumerate(warehouse): 
        for x, val in enumerate(col):
            if val == BOX: 
                total += ((100 * y) + x)
    return total

doctest.run_docstring_examples(calculate_gps_coordinates, None)

def simulate(warehouse, moves): 
    robot = None
    for y, col in enumerate(warehouse): 
        for x, val in enumerate(col):
            if val == ROBOT_CHAR: 
                robot = (y, x)
                break
        if robot: break 
    
    for move in moves: 
        robot = shift(warehouse, robot, move)
        #pprint(move)
        #pprint(warehouse)
        
    
    return calculate_gps_coordinates(warehouse)

warehouse, moves = parse_input(puzzle_input)
print(simulate(warehouse, moves))
```

## Puzzle 2

Challenge of this one is that everything being pushed has to be checked to see if it can be moved, before anything can be moved. 

Our recursive approach only considers one "chain" of motion, so it will fail under these constraints. 

We need to keep track of layers, such that every object in that layer can move. If it can, move it. 
I'm guessing we can still recurse here, given that layers will only be getting wider. that is, if layer 3 can move, then layer 2 can move. The caveat being that in barriers are included in these layers. If a barrier is in a layer, it cannot move

```python
BOX2 = "[]"

test_input = """
#######
#...#.#
#.....#
#..OO@#
#..O..#
#.....#
#######

<vv<<^^<<^^
"""

def scale_input(string): 
    return string.replace(BOX, BOX2).replace(SPACE, SPACE*2).replace(BARRIER, BARRIER*2).replace(ROBOT_CHAR, "@.")

def calculate_gps_coordinates2(warehouse): 
    """
    >>> calculate_gps_coordinates2([list("##########"), list("##...[]..."), list("##........")])
    105
    """

    total = 0 
    for y, col in enumerate(warehouse): 
        for x, val in enumerate(col):
            if val == BOX2[0]: 
                total += ((100 * y) + x)
    return total

doctest.run_docstring_examples(calculate_gps_coordinates2, None)

def check_layer(warehouse, layer, move):
    """    
    >>> warehouse = [list("######"), list("#....#"), list("#[][]#"), list("#..@.#")]
    >>> check_layer(warehouse, [(2, i) for i in range(1,5)], UP)
    True
    
    >>> warehouse = [list("######"), list("#..#.#"), list("#[][]#"), list("#..@.#")]
    >>> check_layer(warehouse, [(2, i) for i in range(1,5)], UP)
    False

    >>> warehouse = [list("##############"), list("##......##..##"), list("##...[][]...##"), list("##....[]....##"), list("##.....@....##")]
    >>> check_layer(warehouse, [(3,6), (3,7)], UP)
    False
    
    >>> warehouse = [list("##############"), list("##.......#..##"), list("##...[][]...##"), list("##....[]....##"), list("##.....@....##")]
    >>> check_layer(warehouse, [(3,6), (3,7)], UP)
    True
    """
    if len(layer) == 0: 
        return True 
    
    dy, dx = MOVES[move]
    next_layer = []
    for y,x in layer: 
        if warehouse[y][x] == BARRIER or warehouse[y+dy][x+dx] == BARRIER: 
            return False 
        if warehouse[y+dy][x+dx] == '[': 
            next_layer.append((y+dy, x+dx))
            next_layer.append((y+dy, x+dx+1))

        if warehouse[y+dy][x+dx] == ']': 
            next_layer.append((y+dy, x+dx))
            next_layer.append((y+dy, x+dx-1))
            
    return check_layer(warehouse, next_layer, move)

doctest.run_docstring_examples(check_layer, None)

```

```python
def shift2(warehouse, push, move):
    """
    # >>> warehouse = [list("#.[]@#")]
    # >>> shift2(warehouse, (0, 4), RIGHT)
    # (0, 4)
    # >>> warehouse  
    # [['#', '.', '[', ']', '@', '#']]
    
    # >>> warehouse = [list("#.[]@#")]
    # >>> shift2(warehouse, (0, 4), LEFT)
    # (0, 3)
    # >>> warehouse  
    # [['#', '[', ']', '@', '.', '#']]

    # >>> warehouse = [list("##############"), list("##......##..##"), list("##...[][]...##"), list("##....[]....##"), list("##.....@....##")]
    # >>> shift2(warehouse, (4, 7), UP)
    # (4, 7)
    # >>> warehouse
    # [['#', '#', '#', '#', '#', '#', '#', '#', '#', '#', '#', '#', '#', '#'], ['#', '#', '.', '.', '.', '.', '.', '.', '#', '#', '.', '.', '#', '#'], ['#', '#', '.', '.', '.', '[', ']', '[', ']', '.', '.', '.', '#', '#'], ['#', '#', '.', '.', '.', '.', '[', ']', '.', '.', '.', '.', '#', '#'], ['#', '#', '.', '.', '.', '.', '.', '@', '.', '.', '.', '.', '#', '#']]
    
    >>> warehouse = [list("##############"), list("##.......#..##"), list("##...[][]...##"), list("##....[]....##"), list("##.....@....##")]
    >>> shift2(warehouse, (4, 7), UP)
    (3, 7)
    >>> warehouse
    [['#', '#', '#', '#', '#', '#', '#', '#', '#', '#', '#', '#', '#', '#'], ['#', '#', '.', '.', '.', '[', ']', '[', ']', '#', '.', '.', '#', '#'], ['#', '#', '.', '.', '.', '.', '[', ']', '.', '.', '.', '.', '#', '#'], ['#', '#', '.', '.', '.', '.', '.', '@', '.', '.', '.', '.', '#', '#'], ['#', '#', '.', '.', '.', '.', '.', '.', '.', '.', '.', '.', '#', '#']]
    """ 
    dy, dx = MOVES[move]
    py, px = push[0], push[1]
    shoved = (py + dy, px + dx)
    sy, sx = shoved[0], shoved[1]
    if warehouse[sy][sx] == BARRIER: # barriers don't move
        return push
    
    if warehouse[sy][sx] in BOX2: 
        if move == LEFT or move == RIGHT: 
            shift2(warehouse, (sy,sx), move)
        else: 
            if warehouse[sy][sx] == '[': 
                if check_layer(warehouse, [(sy,sx), [sy,sx+1]], move): 
                    shift2(warehouse, (sy,sx), move)
                    shift2(warehouse, (sy,sx+1), move)
            if warehouse[sy][sx] == ']': 
                if check_layer(warehouse, [(sy,sx-1), [sy,sx]], move): 
                    shift2(warehouse, (sy,sx-1), move)
                    shift2(warehouse, (sy,sx), move)
    
    if warehouse[sy][sx] == SPACE: # open, can just shift
        warehouse[sy][sx], warehouse[py][px] =  warehouse[py][px], warehouse[sy][sx]
        return (sy, sx)

    return push

doctest.run_docstring_examples(shift2, None)
```
