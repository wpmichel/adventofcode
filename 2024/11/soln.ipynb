```python
import doctest

puzzle_input = """5178527 8525 22 376299 3 69312 0 275"""
NEWLINE = '\n'

def parse_input(string: str): 
    string = string.strip(NEWLINE)
    return list(map(int, string.split(' ')))

doctest.run_docstring_examples(parse_input, None)
 
from heapq import heappop, heappush


def transform_stone(stone):
    if stone == 0:
        return stone
    else:
        stone_str = str(stone)
        if len(stone_str) % 2 == 0:
            mid = len(stone_str) // 2
            left = int(stone_str[:mid])
            right = int(stone_str[mid:])
            return (left, right)
        else:
            return stone * 2024
        
def blinks(stones, blinks):
    """Process all stones for the given number of blinks, splitting large sets."""
    total = 0
    heap = []
    iters = 0
    for stone in stones:
        heappush(heap, (blinks, stone))
    while heap: 
        blinks, stone = heappop(heap)
        if blinks == 0: 
            total += 1
            continue
        else: 
            if stone == 0:
                heappush(heap, (blinks-1, 1))
            else:
                stone_str = str(stone)
                if len(stone_str) % 2 == 0:
                    mid = len(stone_str) // 2
                    left = int(stone_str[:mid])
                    right = int(stone_str[mid:])
                    heappush(heap, (blinks-1, left))
                    heappush(heap, (blinks-1, right))
                else: 
                    heappush(heap, (blinks-1, stone*2024))    
        iters += 1
        if iters == 100_000: 
            print(f"{len(heap)} @ {iters}")
    return total

doctest.run_docstring_examples(blink, None)
stones = parse_input("125")
print(blinks(stones, 25))
```

```python
def transform_stone(stone):
    """Calculate immediate transformations of a stone."""
    if stone == 0:
        return [1]
    stone_str = str(stone)
    if len(stone_str) % 2 == 0:
        mid = len(stone_str) // 2
        return [int(stone_str[:mid]), int(stone_str[mid:])]
    return [stone * 2024]

def count_stones_dp(initial_stones, max_steps):
    """
    Count stones using bottom-up dynamic programming.
    dp[stone][step] = number of stones this will become after 'step' steps
    """
    dp = {}
    
    def ensure_stone_in_dp(stone):
        """Ensure stone exists in DP table with base case."""
        if stone not in dp:
            dp[stone] = {0: 1}  # Base case
            
    def calculate_step(stone, step):
        """Calculate result for a specific step if not already known."""
        ensure_stone_in_dp(stone)
        if step not in dp[stone]:
            # Get next stones after transformation
            next_stones = transform_stone(stone)
            # Ensure all next stones are initialized
            for next_stone in next_stones:
                ensure_stone_in_dp(next_stone)
            # Calculate previous step first if needed
            if step > 1:
                for s in range(1, step):
                    calculate_step(stone, s)
            # Calculate this step
            dp[stone][step] = sum(calculate_step(next_stone, step - 1) 
                                for next_stone in next_stones)
        return dp[stone][step]
    
    # Process each initial stone
    total = 0
    for stone in initial_stones:
        result = calculate_step(stone, max_steps)
        total += result
        
    return total

# Example usage and testing
if __name__ == '__main__':
    stones = parse_input(puzzle_input)
    steps = 75
    print(count_stones_dp(stones, steps))
    
```
