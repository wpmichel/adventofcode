```python
import doctest

NEWLINE = '\n'

def parse_input(string): 
    calibrations = []
    for line in string.strip(NEWLINE).split(NEWLINE): 
        calibration, vals = line.split(': ')
        calibrations.append((int(calibration), list(map(int, vals.split(' ')))))
    return calibrations

def process_vals(vals, ops):
    """
    >>> process_vals([1,2], ['+'])
    3
    >>> process_vals([1,2,3], ['+', '*'])
    9
    >>> process_vals([1,2], ['||'])
    12
    >>> process_vals([1,2,1], ['||', '+'])
    13
    """ 
    total = vals[0]
    for idx, op in enumerate(ops):
        if op == '+': 
            total += vals[idx+1]
        elif op == '*':
            total *= vals[idx+1]
        else: 
            total = int(f"{total}{vals[idx+1]}")
    return total

doctest.run_docstring_examples(process_vals, None)
```
