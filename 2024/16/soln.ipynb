# Puzzle 1
Seems simple. DFS shortest path. Wrinkle is the change of direction. 

```python
test_input = """
###############
#.......#....E#
#.#.###.#.###.#
#.....#.#...#.#
#.###.#####.#.#
#.#.#.......#.#
#.#.#####.###.#
#...........#.#
###.#.#####.#.#
#...#.....#.#.#
#.#.#.###.#.#.#
#.....#...#.#.#
#.###.#.#.#.#.#
#S..#.....#...#
###############
"""


from heapq import heappush, heappop
from collections import defaultdict


def in_bounds(arr, y, x): 
    return 0 <= y < len(arr) and 0 <= x < len(arr[0])

def get_neighbors(arr, pos): 
    neighbors = set()
    for offset in [(-1,0), (1, 0), (0, 1), (0, -1)]: 
        ny, nx = pos[0]+offset[0], pos[1]+offset[1]
        if in_bounds(arr, ny, nx) and arr[ny][nx] != '#':
            neighbors.add((ny, nx))
    return neighbors

def get_direction(prev, curr):
    # Return direction as number: 0=North, 1=East, 2=South, 3=West
    dy, dx = curr[0] - prev[0], curr[1] - prev[1]
    if dy == -1: return 0  # North
    if dx == 1: return 1   # East
    if dy == 1: return 2   # South
    if dx == -1: return 3  # West
    return None

def direction_change_cost(prev_pos, curr_pos, next_pos):
    if not prev_pos: return 0
    prev_dir = get_direction(prev_pos, curr_pos)
    next_dir = get_direction(curr_pos, next_pos)
    return 1000 if prev_dir != next_dir else 0


def shortest_path(arr, start, end): 
    visited = defaultdict(lambda: float('inf'))
    # (cost, pos, last pos, path)
    queue = [(0, start, (start[0], start[1]-1), [(start[0], start[1]-1), start])] # start facing east, so act like we have moved from the left

    
    while queue: 
        cost, curr, prev, path = heappop(queue)

        if curr == end: 
            return cost, path
        
        for next_pos in get_neighbors(arr, curr): 
            if next_pos == prev: 
                continue
            
            direction = get_direction(curr, next_pos)
            next_cost = cost + 1 + direction_change_cost(prev, curr, next_pos)
            state = (next_pos, direction)
            if next_cost < visited[state]: 
                visited[state] = next_cost
                heappush(queue, (next_cost, next_pos, curr, path + [next_pos]))

    
    return float('inf'), []

arr = parse_input(puzzle_input)
start, end = None, None
for y, row in enumerate(arr): 
    for x, val in enumerate(row): 
        if val == 'S': 
            start = (y, x)
        if val == 'E': 
            end = (y, x)
print(shortest_path(arr, start, end))
```

# Puzzle 2 
Well, we have an efficient way to find an best path, but it prunes other potential paths. 
Searching every path is going to be too slow. Maybe it can be fast enough if we 
1. Prune any path longer than the best
2. Skip and point that's already part of a best path. 


My intuition for 2 might be incorrect. Maybe it's possible that there are two equally short paths from a given point. something like 

#######
##OEO##
##O#O##
##OOO##
###O###
###S###
#######

so let's try with 1 and see how far it gets us

```python
test_input2 = """
#################
#...#...#...#..E#
#.#.#.#.#.#.#.#.#
#.#.#.#...#...#.#
#.#.#.#.###.#.#.#
#...#.#.#.....#.#
#.#.#.#.#.#####.#
#.#...#.#.#.....#
#.#.#####.#.###.#
#.#.#.......#...#
#.#.###.#####.###
#.#.#...#.....#.#
#.#.#.#####.###.#
#.#.#.........#.#
#.#.#.#########.#
#S#.............#
#################
# """

def nodes_in_shortest_paths(arr, start, end, min_cost):
    optimal_paths = set()
    
    def shortest_path(arr, start, end, min_cost, optimal_paths): 
        visited = defaultdict(lambda: float('inf'))
        # (cost, pos, last pos, path)
        queue = [(0, start, (start[0], start[1]-1), {(start[0], start[1]-1), start})] # start facing east, so act like we have moved from the left

        while queue: 
            cost, curr, prev, path = heappop(queue)

            if curr == end: 
                optimal_paths |= path
                continue        
    
            for next_pos in get_neighbors(arr, curr): 
                if next_pos == prev: 
                    continue
                
                direction = get_direction(curr, next_pos)
                next_cost = cost + 1 + direction_change_cost(prev, curr, next_pos)
                if next_cost > min_cost: 
                    continue 
                
                state = (next_pos, direction)
                if next_cost <= visited[state]: 
                    visited[state] = next_cost
                    heappush(queue, (next_cost, next_pos, curr, path | {next_pos}))

    
    shortest_path(arr, start, end, min_cost, optimal_paths)
    return optimal_paths

arr = parse_input(puzzle_input)
start, end = None, None
for y, row in enumerate(arr): 
    for x, val in enumerate(row): 
        if val == 'S': 
            start = (y, x)
        if val == 'E': 
            end = (y, x)


min_cost, _ = shortest_path(arr, start, end)
print(min_cost)
visited = nodes_in_shortest_paths(arr, start, end, min_cost)
visited.remove(start)
print(len(visited))
```
